```python
from skimage.measure import compare_ssim as ssim
#import skimage
#print(skimage.__version__)
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
from PIL import Image
```

```python
images = [] 
address_name = '/home/hsaeidi/Pictures/r01/'
for j in range(1,4):
    tmp = address_name + str((j)) +'.png'
    images.append(tmp)
images_no = len(images)
print('Total number of image addresses: ', images_no)


img1 = Image.open(str(images[0]))
img2 = Image.open(str(images[1]))
img3 = Image.open(str(images[2]))

print('The sample image size is')
print(img1.size)
#print(' and it contains this type of data')
#print(np.asarray(img1))
#print('when normalized it gets like this')
#print(np.asarray(img1)/float(255))

img1 = np.asarray(img1)
img2 = np.asarray(img2)
img3 = np.asarray(img3)
print(img1.shape)

#plot 2 of them
fig = plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(img1)
plt.title('img 1')
plt.subplot(132)
plt.imshow(img2)
plt.title('img 2')
plt.subplot(133)
plt.imshow(img3)
plt.title('img 3')

(score, diff) = ssim(img1, img3,full=True, multichannel=True)
print(score)
```
